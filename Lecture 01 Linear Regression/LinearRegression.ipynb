```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv(r'C:\Users\Aman\Desktop\Machine Learning\Lecture 01 Linear Regression\dataset\area_price.csv')
df.head()
```

```python
df.columns
```

```python
plt.title('Relation between Area (sq ft) vs Price ($)')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price ($)')
# plt.plot(df['Area (sq ft)'], df['Price ($)'],  marker='o',  linestyle='--', color='blue')
plt.scatter(df['Area (sq ft)'], df['Price ($)'], color='red', marker='+')
plt.show()

```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['Area (sq ft)']], df[['Price ($)']])
```

```python
reg.predict([[800]])
```

```python
newdf = pd.read_csv('dataset/canada_per_capita_income.csv')
```

```python
newdf
```

```python
newdf.columns
```

```python
model = linear_model.LinearRegression()
```

```python
model.fit(newdf[['year']], newdf[['per capita income (US$)']])
```

```python
model.predict([[2020]])
```

```python
plt.title('Relation between year and per capita income (US$)')
plt.xlabel('year')
plt.ylabel('capita income (US$)')

# Scatter plot for actual data with a label
plt.scatter(newdf['year'], newdf['per capita income (US$)'], color='red', marker='o', label='Actual Data')

# Line plot for predicted data with a label
plt.plot(newdf['year'], model.predict(newdf[['year']]), color='blue', linestyle='-', label='Predicted Data')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
